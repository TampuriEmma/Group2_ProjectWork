### Replace nan in postal code using count function

```python
## check for the count of each postal code
super_store_df['Postal Code'].value_counts()
```

```python
## select the index for the highest count
super_store_df['Postal Code'].value_counts().index[0]
```

```python
## replace the nans with the highest counts of postal code
super_store_df['Postal Code'] = super_store_df['Postal Code'].fillna(super_store_df['Postal Code'].value_counts().index[0])
```

### Check for NaN in our data

```python
super_store_df['Postal Code'].isnull().sum()
```

## Data Visualization

```python
plt.scatter(super_store_df['Shipping Cost'], super_store_df['Sales'],c="red")

plt.title('Relation between Shipping Cost and Sales')

plt.ylabel('Sales')

plt.xlabel('Shipping Cost')

plt.savefig('shipping_cost_sales_rel.png')

plt.show
```

```python
my_plot = sns.countplot(x=super_store_df['Market'],color="green")
my_plot.set_xticklabels(my_plot.get_xticklabels(),rotation=90)
plt.title('Destribtion of Market')
plt.grid(axis='y')
plt.xlabel('Markets')

plt.savefig('markets_counts.png')

plt.show
```

```python
sns.countplot(x=super_store_df['Ship Mode'],color="green")
plt.title('Destribtion of Shipping Mode')
plt.grid(axis='y')
plt.xlabel('Shipping Mode')

plt.savefig('Shipping_Mode_counts.png')

plt.show
```

```python
sns.barplot(super_store_df, y='Shipping Cost', x='Ship Mode', errorbar=None)

plt.grid(axis='y')
plt.title("Shipping Mode that Cost More")
plt.xlabel('Shipping Mode')
plt.ylabel('Shipping Cost (Dollars)')


plt.show
```

```python
sns.countplot(super_store_df, x='Ship Mode', hue='Order Priority')


plt.title("Relationship between the Shipping Mode of Prodects and Order Priority")
plt.xlabel('Shipping Mode')
plt.grid(True,axis='y')

plt.show
```

```python
plt.figure(figsize=(16,8))
sns.barplot(super_store_df, y='Sales', x='Sub-Category', errorbar=None)

plt.title("Relationship between the Sub-Category of Prodects and Profit")
plt.xlabel('Product Sub-Category')
plt.ylabel('Profit (Dollars)')
plt.grid(True,axis='y')

plt.show
```

```python
plt.figure(figsize=(16,8))
sns.barplot(super_store_df, y='Profit', x='Sub-Category', errorbar=None)

plt.title("Relationship between the Sub-Category of Prodects and Profit")
plt.xlabel('Product Category')
plt.ylabel('Profit (Dollars)')
plt.grid(True,axis='y')

plt.show
```

```python
plt.figure(figsize=(16,8))
sns.barplot(super_store_df, y='Discount', x='Sub-Category', errorbar=None,color = 'green')

plt.title("Relationship between the Sub-Category of Prodects and Discount")
plt.xlabel('Product Sub-Category')
plt.ylabel('Discounts')
plt.grid(True,axis='y')

plt.show
```

```python
plt.figure(figsize=(18,8))
sns.barplot(super_store_df, x='Sub-Category', y='Quantity', errorbar=None, color='brown')


plt.title("Relationship between Sub-Category of Prodects and Quantity")

plt.grid(True,axis='y')

plt.show
```

```python
plt.figure(figsize=(10,10))
sub_category_count = super_store_df['Sub-Category'].value_counts()
plt.pie(sub_category_count, labels = sub_category_count.index ,autopct='%1.1f%%')
plt.title("Destribution of Sub-Category of Products")
plt.show()
```
